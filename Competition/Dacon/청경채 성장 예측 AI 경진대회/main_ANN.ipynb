```python
cols_convert_info = {
    "Ïô∏Î∂ÄÏò®ÎèÑÏ∂îÏ†ïÍ¥ÄÏ∏°Ïπò": "Ïô∏Î∂ÄÏò®ÎèÑÍ¥ÄÏ∏°Ïπò",
    "Ïô∏Î∂ÄÏäµÎèÑÏ∂îÏ†ïÍ¥ÄÏ∏°Ïπò": "Ïô∏Î∂ÄÏäµÎèÑÍ¥ÄÏ∏°Ïπò",
    "ÌéåÌîÑÏµúÍ∑ºÎ∂ÑÎ¨¥Îüâ": "ÏµúÍ∑ºÎ∂ÑÎ¨¥Îüâ",
    "ÌéåÌîÑÏùºÍ∞ÑÎàÑÏ†ÅÎ∂ÑÎ¨¥Îüâ": "ÏùºÍ∞ÑÎàÑÏ†ÅÎ∂ÑÎ¨¥Îüâ",
    "ÌéåÌîÑÏûëÎèôÎÇ®ÏùÄÏãúÍ∞Ñ.1": "ÌéåÌîÑÏûëÎèôÎÇ®ÏùÄÏãúÍ∞Ñ"
}

col_list = [
    'ÏãúÍ∞Ñ', 'ÎÇ¥Î∂ÄÏò®ÎèÑÍ¥ÄÏ∏°Ïπò', 'ÎÇ¥Î∂ÄÏäµÎèÑÍ¥ÄÏ∏°Ïπò', 'CO2Í¥ÄÏ∏°Ïπò', 'ECÍ¥ÄÏ∏°Ïπò', 'Ïô∏Î∂ÄÏò®ÎèÑÍ¥ÄÏ∏°Ïπò', 'Ïô∏Î∂ÄÏäµÎèÑÍ¥ÄÏ∏°Ïπò', 'ÌéåÌîÑÏÉÅÌÉú', 'ÌéåÌîÑÏûëÎèôÎÇ®ÏùÄÏãúÍ∞Ñ', 'ÏµúÍ∑ºÎ∂ÑÎ¨¥Îüâ', 'ÏùºÍ∞ÑÎàÑÏ†ÅÎ∂ÑÎ¨¥Îüâ', 'ÎÉâÎ∞©ÏÉÅÌÉú', 'ÎÉâÎ∞©ÏûëÎèôÎÇ®ÏùÄÏãúÍ∞Ñ', 'ÎÇúÎ∞©ÏÉÅÌÉú', 'ÎÇúÎ∞©ÏûëÎèôÎÇ®ÏùÄÏãúÍ∞Ñ', 'ÎÇ¥Î∂ÄÏú†ÎèôÌå¨ÏÉÅÌÉú', 'ÎÇ¥Î∂ÄÏú†ÎèôÌå¨ÏûëÎèôÎÇ®ÏùÄÏãúÍ∞Ñ', 'Ïô∏Î∂ÄÌôòÍ∏∞Ìå¨ÏÉÅÌÉú', 'Ïô∏Î∂ÄÌôòÍ∏∞Ìå¨ÏûëÎèôÎÇ®ÏùÄÏãúÍ∞Ñ', 'ÌôîÏù¥Ìä∏ LEDÏÉÅÌÉú', 'ÌôîÏù¥Ìä∏ LEDÏûëÎèôÎÇ®ÏùÄÏãúÍ∞Ñ', 'ÌôîÏù¥Ìä∏ LEDÎèôÏûëÍ∞ïÎèÑ', 'Î†àÎìú LEDÏÉÅÌÉú', 'Î†àÎìú LEDÏûëÎèôÎÇ®ÏùÄÏãúÍ∞Ñ', 'Î†àÎìú LEDÎèôÏûëÍ∞ïÎèÑ', 'Î∏îÎ£® LEDÏÉÅÌÉú', 'Î∏îÎ£® LEDÏûëÎèôÎÇ®ÏùÄÏãúÍ∞Ñ', 'Î∏îÎ£® LEDÎèôÏûëÍ∞ïÎèÑ', 'Ïπ¥Î©îÎùºÏÉÅÌÉú', 'ÎÉâÎ∞©Ïò®ÎèÑ', 'ÎÇúÎ∞©Ïò®ÎèÑ', 'Í∏∞Ï§ÄÏò®ÎèÑ', 'ÎÇúÎ∞©Î∂ÄÌïò', 'ÎÉâÎ∞©Î∂ÄÌïò', 'Ï¥ùÏ∂îÏ†ïÍ¥ëÎüâ', 'Î∞±ÏÉâÍ¥ëÏ∂îÏ†ïÍ¥ëÎüâ', 'Ï†ÅÏÉâÍ¥ëÏ∂îÏ†ïÍ¥ëÎüâ', 'Ï≤≠ÏÉâÍ¥ëÏ∂îÏ†ïÍ¥ëÎüâ', 'id'
]

class feature_preprocessor_v0():
    def __init__(self, id_value=None, sampling_unit=10, date_var="ÏãúÍ∞Ñ"):
        self.id_value = id_value
        self.sampling_unit = sampling_unit
        self.date_vars = date_var
    def fit(self, x):
        pass
    def transform(self, x):   
        x_copy = copy.deepcopy(x)


        for i in cols_convert_info.keys():
            if (i in x_copy.columns) & (cols_convert_info[i] in x_copy.columns):
                x_copy.drop
                tmp1 = x_copy[cols_convert_info[i]]
                tmp2 = x_copy[i]
                tmp2[tmp1 != 0.0] = 0
                tmp2 = tmp2.fillna(0)
                x_copy[cols_convert_info[i]] = tmp1 + tmp2
                x_copy = x_copy.drop(i, axis=1)
            elif i in x_copy.columns:
                tmp_cols = series(x_copy.columns)
                tmp_cols[x_copy.columns == i] = cols_convert_info[i]
                x_copy.columns = tmp_cols

        x_copy[x_copy.select_dtypes("number").columns] = x_copy.select_dtypes("number").interpolate(method="linear").bfill()
        x_copy = x_copy.iloc[::self.sampling_unit, :].reset_index(drop=True)
        x_copy.iloc[:, 1:] = x_copy.iloc[:, 1:].astype("float32")
        x_copy["id"] = self.id_value

        x_copy.columns = [i.strip() for i in list(x_copy.columns)]
        x_copy = x_copy[col_list]

        return x_copy
```

```python
def get_dataset(input_paths, target_paths=None, sampling_unit=15, infer=False):
    data_list = []
    label_list = []
    print('Data Pre-processing..')
    for input_path, target_path in tqdm(zip(input_paths, target_paths)):
        input_df = pd.read_csv(input_path, parse_dates=["ÏãúÍ∞Ñ"])
        target_df = pd.read_csv(target_path)

        # if na values are more than 20%, drop
        if not infer:
            if input_df.iloc[:, 1:].isna().sum().sum() > ((input_df.shape[0] * input_df.shape[1] - 1) * 0.2):
                print("CONTINUE : too many NAs-", input_path.split("/")[-1])
                continue

        # ÌäπÏ†ï Ïª¨ÎüºÏù¥ ÏóÜÎäî timeseriesÎäî drop
        try:
            fp = feature_preprocessor_v0(id_value=input_path.split("/")[-1], sampling_unit=sampling_unit)
            fp.fit(input_df)
            input_df = fp.transform(input_df)
        except:
            continue

        for idx, value in enumerate(target_df["rate"]):
            time_series = input_df[int((1440 / sampling_unit)*idx) : int((1440 / sampling_unit)*(idx+1))].reset_index(drop=True)
            if not infer:
                if time_series.isna().sum().sum() > 0:
                    print("ERROR : NA -", input_path.split("/")[-1])
                    continue
            data_list.append(time_series)
            label_list.append(value)
    return data_list, label_list
```

```python
from calendar import monthrange

def ftransformer_timestamp(X, date_var):
    X_copy = copy.deepcopy(X)

    X_copy["quarter"] = X[date_var].dt.quarter
    X_copy["month"] = X[date_var].dt.month
    X_copy["week_of_month"] = X[date_var].apply(week_of_month)

    X_copy["weekday"] = X[date_var].dt.weekday
    X_copy["holiday"] = X_copy["weekday"].apply(lambda x: 1 if x in (5, 6) else 0)
    X_copy["hour"] = X[date_var].dt.hour
    # X_copy["hour_split4"] = pd.cut(X_copy["hour"], bins=[-np.inf, 6, 12, 18, 24, np.inf], labels=False, right=False)
    X_copy["hour_split8"] = pd.cut(X_copy["hour"], bins=[-np.inf, 3, 6, 9, 12, 15, 18, 21, 24, np.inf], labels=False, right=False)
    X_copy["office_hour"] = X_copy["hour"].apply(lambda x: 1 if (x >= 9 and x <= 18) else 0)

    X_copy["secs_to_year"] = X[date_var].apply(lambda x: (x - x.replace(month=1, day=1, hour=0, minute=0, second=0, microsecond=0)).total_seconds() / (365 * 24 * 3600)).astype("float32")
    X_copy["secs_to_month"] = X[date_var].apply(lambda x: (x - x.replace(day=1, hour=0, minute=0, second=0, microsecond=0)).total_seconds() / (monthrange(x.year, x.month)[1] * 24 * 3600)).astype("float32")
    X_copy["secs_to_week"] = X[date_var].apply(lambda x: (x - (x - timedelta(days=x.weekday())).replace(hour=0, minute=0, second=0, microsecond=0)).total_seconds() / (7 * 24 * 3600)).astype("float32")
    X_copy["secs_to_day"] = X[date_var].apply(lambda x: (x - x.replace(hour=0, minute=0, second=0, microsecond=0)).total_seconds() / (1 * 24 * 3600)).astype("float32")
    return X_copy

def transform_timeseries_data(x):
    x = pd.concat([x.drop("quarter", axis=1), dataframe(tf.keras.utils.to_categorical(x[["quarter"]]-1, 4), columns=["quarter_" + str(i) for i in range(4)])], axis=1)
    x = pd.concat([x.drop("month", axis=1), dataframe(tf.keras.utils.to_categorical(x[["month"]]-1, 12), columns=["month_" + str(i) for i in range(12)])], axis=1)
    x = pd.concat([x.drop("week_of_month", axis=1), dataframe(tf.keras.utils.to_categorical(x[["week_of_month"]]-1, 5), columns=["week_of_month_" + str(i) for i in range(5)])], axis=1)
    x = pd.concat([x.drop("weekday", axis=1), dataframe(tf.keras.utils.to_categorical(x[["weekday"]], 7), columns=["weekday_" + str(i) for i in range(7)])], axis=1)
    # x = pd.concat([x.drop("hour", axis=1), dataframe(tf.keras.utils.to_categorical(x[["hour"]], 24), columns=["hour_" + str(i) for i in range(24)])], axis=1)
    # x = pd.concat([x.drop("hour_split4", axis=1), dataframe(tf.keras.utils.to_categorical(x[["hour_split4"]], 4), columns=["hour_split4_" + str(i) for i in range(4)])], axis=1)
    x = pd.concat([x.drop("hour_split8", axis=1), dataframe(tf.keras.utils.to_categorical(x[["hour_split8"]], 8), columns=["hour_split8_" + str(i) for i in range(8)])], axis=1)
    return x.drop("hour", axis=1)

class feature_engineering_v0():
    def __init__(self, date_var="ÏãúÍ∞Ñ"):
        self.date_var = date_var
    def fit(self, x):
        pass
    def transform(self, x):
        x_copy = copy.deepcopy(x)
        return x_copy

class feature_engineering_v1():
    def __init__(self, date_var="ÏãúÍ∞Ñ"):
        self.date_var = date_var
    def fit(self, x):
        pass
    def transform(self, x):
        x_copy = copy.deepcopy(x)
        x_copy[self.date_var] = x_copy[self.date_var].bfill().ffill()
        # x_copy[x_copy.select_dtypes("number").columns] = x_copy.select_dtypes("number").interpolate(method="linear").bfill()
        # x_copy = x_copy.iloc[::self.sampling_unit, :].reset_index(drop=True)
        x_copy = pd.merge(x_copy, ftransformer_timestamp(x_copy[[self.date_var]], self.date_var), how="left", on=self.date_var)
        # x_copy.iloc[:, 1:] = x_copy.iloc[:, 1:].astype("float32")

        # x_copy.columns = [i.strip() for i in list(x_copy.columns)]
        # x_copy = x_copy[col_list]

        # x_copy["id"] = self.id_value
        return x_copy
```

```python
def create_dataset(x, y, batch_size, shuffle=True, sample_weight=None):
    if sample_weight is not None:
        dataset = tf.data.Dataset.from_tensor_slices((x, y, sample_weight))
    else:
        dataset = tf.data.Dataset.from_tensor_slices((x, y))
    dataset = dataset.shuffle(int(batch_size * 4), reshuffle_each_iteration=True) if shuffle else dataset
    dataset = dataset.batch(batch_size)
    dataset = dataset.prefetch(tf.data.AUTOTUNE)
    return dataset
```

```python
dropoutRate = 0.5

# LSTM Dense
def create_model(input_shape=(96, 100)):
    input_tensor = layers.Input(shape=input_shape, dtype=tf.float32)

    x = layers.LSTM(64, dropout=0.25, return_sequences=True)(input_tensor)
    x = layers.LSTM(16, dropout=0.25, return_sequences=True)(x)
    x = layers.Flatten()(x)
    
    x = layers.Dropout(dropoutRate)(x)
    x = layers.Dense(1024, activation="relu")(x)
    x = layers.Dropout(dropoutRate)(x)
    last_layer = layers.Dense(256, activation="relu")(x)

    regressor = layers.Dense(1)(last_layer)

    return Model(input_tensor, regressor)
```

```python
def do_fold_training(fold, train_idx, valid_idx, sample_weight=None, finetuning=False, scaling=False):
    tmp_time = time()
    print("\n===== Fold", fold, "=====\n")
    global val_pred, test_pred

    cb_tensorboard = tf_callbacks.TensorBoard(log_dir=tb_filepath, histogram_freq=1)
    try:
        wandb.tensorboard.patch(root_logdir=tb_filepath)
    except:
        pass
    wandb.init(
        project="dacon_leaf_weight_rete_prediction",
        group=architecture_name + "_fpV" + str(feature_preprocessor_version) + "_feV" + str(feature_version),
        name="fold_" + str(fold)
    ); wandb.config.step = 0

    train_x = df_full_x[train_idx]
    valid_x = df_full_x[valid_idx]
    train_y = df_full_y[train_idx]
    valid_y = df_full_y[valid_idx]
    test_x = df_test_x
    input_shape = train_x.shape

    if scaling:
        scaler = MinMaxScaler()
        scaler.fit(np.concatenate([i for i in train_x], axis=0))

        train_x = np.concatenate([scaler.transform(i)[np.newaxis, ...] for i in train_x], axis=0)
        valid_x = np.concatenate([scaler.transform(i)[np.newaxis, ...] for i in valid_x], axis=0)
        test_x = np.concatenate([scaler.transform(i)[np.newaxis, ...] for i in test_x], axis=0)
        scaler_list.append(scaler)

    train_ds = create_dataset(train_x, train_y, sample_weight=sample_weight, batch_size=batch_size, shuffle=True)
    valid_ds = create_dataset(valid_x, valid_y, sample_weight=None, batch_size=batch_size, shuffle=False)
    
    total_iter = int(np.ceil(input_shape[0] / batch_size) * epochs)
    SCHEDULE_BOUNDARIES = [int(total_iter*0.2), int(total_iter*0.4), int(total_iter*0.6), int(total_iter*0.8)]
    lr_schedule = tf.keras.optimizers.schedules.PiecewiseConstantDecay(
        boundaries=SCHEDULE_BOUNDARIES,
        values=[5e-4, 1e-4, 5e-5, 1e-5, 5e-6],
    )

    cb_earlyStopping = tf_callbacks.EarlyStopping(
        patience=patient_epochs, monitor='val_tf_losses_rmse', mode='min'
    )
    cb_modelsave = tf_callbacks.ModelCheckpoint(
        filepath=checkpoint_filepath, monitor='val_tf_losses_rmse', mode='min', save_weights_only=True, save_best_only=True
    )

    model = create_model(input_shape[1:])
    print(model.summary())
    model.compile(
        loss=tf.keras.losses.MeanSquaredError(name="loss"),
        optimizer=tfa.optimizers.AdamW(learning_rate=lr_schedule, weight_decay=weight_decay),
        metrics=[tf_losses_rmse]
    )
    
    print("start training")
    model.fit(train_ds, validation_data=valid_ds, epochs=epochs, verbose=0,
        callbacks=[cb_modelsave, cb_earlyStopping, cb_tensorboard, TqdmCallback(verbose=1)])
    print("end training")

    model.load_weights(checkpoint_filepath)
    model.trainable = False
    

    fold_metric.append(model.evaluate(valid_ds, return_dict=True))
    print("fold", fold, "score :", fold_metric[-1])

    valid_pred[valid_idx] = model.predict(valid_ds)
    test_pred += model.predict(test_x) / n_folds

    wandb.log({
        "fold": fold,
        "mae": metrics.mean_absolute_error(valid_y, valid_pred[valid_idx, 0]),
        "rmse": metrics.mean_squared_error(valid_y, valid_pred[valid_idx, 0], squared=False)
    })
    wandb.finish()

    model_list.append(model)
 
    print("fold", fold, "time to training :", round(time() - tmp_time, 3))
```

```python
# learning parameter setting
epochs = 30
patient_epochs = 10
batch_size = 8

eta = 1e-3 # deprecated : lr is replaced by scheduler 
weight_decay = 1e-4

checkpoint_filepath = './tmp_checkpoint/'
tb_filepath = "./tmp_tb/"
```

```python
model_list = []
scaler_list = []
fold_metric = []

valid_pred = np.zeros(shape=(df_full_x.shape[0], 1))
test_pred = np.zeros(shape=(df_test_x.shape[0], 1))

n_folds = 10
kfolds_spliter = StratifiedKFold(n_folds, shuffle=True, random_state=42)
target_discretizer = KBinsDiscretizer(n_folds, encode="ordinal", strategy="uniform")
stracVec = target_discretizer.fit_transform(df_full_y.reshape(-1, 1)).flatten()

start_time_training = time()
# fold training
seed_everything()
for fold, (train_idx, valid_idx) in enumerate(kfolds_spliter.split(df_full_x, stracVec)):
    tf.keras.backend.clear_session()
    start_mem = memory_usage()
    do_fold_training(fold, train_idx, valid_idx, sample_weight=None, finetuning=False, scaling=True)
    gc.collect()
    end_mem = memory_usage()
    print("@Memory leaked :", end_mem - start_mem, "\n")
end_time_training = time()
```
