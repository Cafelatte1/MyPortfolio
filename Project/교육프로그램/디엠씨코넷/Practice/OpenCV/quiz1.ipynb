## Setup

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img1 = cv2.cvtColor(cv2.imread(r"C:\Users\Admin\Desktop\class\lecture2_image_processing\quiz\Q1\airplane.bmp"), cv2.COLOR_BGR2RGB)
(img1_h, img1_w) = img1.shape[:2]
print(f"width : {img1_w}, height : {img1_h}")
plt.imshow(img1); plt.show()
```

```python
img2 = cv2.cvtColor(cv2.imread(r"C:\Users\Admin\Desktop\class\lecture2_image_processing\quiz\Q1\mask_plane.bmp"), cv2.COLOR_BGR2RGB)
(img2_h, img2_w) = img2.shape[:2]
print(f"width : {img2_w}, height : {img2_h}")
plt.imshow(img2); plt.show()
```

```python
# resizing

img1 = cv2.resize(img1, (img1_w * 2, img1_h * 2))
(img1_h, img1_w) = img1.shape[:2]
print(f"image1 width : {img1_w}, height : {img1_h}")

img2 = cv2.resize(img2, (img2_w * 2, img2_h * 2))
(img2_h, img2_w) = img2.shape[:2]
print(f"image2 width : {img2_w}, height : {img2_h}")
```

```python
# rotating

img1_rot45 = cv2.warpAffine(
    img1,
    cv2.getRotationMatrix2D((img1_w // 2, img1_h // 2), 45, 1.0),
    (img1_w, img1_h)
)
plt.imshow(img1_rot45); plt.show()

img2_rot45 = cv2.warpAffine(
    img2,
    cv2.getRotationMatrix2D((img2_w // 2, img2_h // 2), 45, 1.0),
    (img2_w, img2_h)
)
plt.imshow(img2_rot45); plt.show()
```

```python
img_onlyAirplane = cv2.copyTo(img1_rot45, img2_rot45, img2_rot45)
plt.imshow(img_onlyAirplane); plt.show()
```
