## Setup

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img1 = cv2.cvtColor(cv2.imread(r"C:\Users\Admin\Desktop\class\lecture2_image_processing\quiz\Q2\candies2.png"), cv2.COLOR_BGR2RGB)
(img1_h, img1_w) = img1.shape[:2]
print(f"width : {img1_w}, height : {img1_h}")
plt.imshow(img1); plt.show()
```

```python
# covert RGB to HSV
img1 = cv2.cvtColor(img1, cv2.COLOR_RGB2HSV)
```

```python
hue_range = {
    "blue": (np.array([100, 100, 100]), np.array([150, 255, 255])),
    "orange": (np.array([5, 0, 0]), np.array([12, 255, 255])),
}
```

```python
# Create a mask for the blue color.
blue_mask = cv2.inRange(img1[:, :, 0], 100, 150)
# Apply the mask to the image.
image_blue = cv2.bitwise_and(img1, img1, mask=blue_mask)
plt.imshow(cv2.cvtColor(image_blue, cv2.COLOR_HSV2RGB)); plt.show()
```

```python
# Create a mask for the blue color.
orange_mask = cv2.inRange(img1[:, :, 0], 0, 20)
# Apply the mask to the image.
image_orange = cv2.bitwise_and(img1, img1, mask=orange_mask)
plt.imshow(cv2.cvtColor(image_orange, cv2.COLOR_HSV2RGB)); plt.show()
```

```python
# add two images
img2 = image_blue + image_orange
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_HSV2RGB)); plt.show()
```
